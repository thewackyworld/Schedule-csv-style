```python
count =0;
while count != len(data):
    if pd.notnull(data.loc[count, 'Course']):
        newC = data.Course[count];
    else:
        data.Course[count] = newC
    count = count +1;


data
```

```python

newdat= data['Days and Times'].str.split(' ', expand=True)
time = newdat[1].str.split('-', expand=True)
newdat[1] = time[0]
newdat[2] = time[1]
print(newdat)
size = data.shape[0]
size
```

```python
SCOPES = ['https://www.googleapis.com/auth/calendar']

def main(i, data, newdat):
            creds = None
            if os.path.exists('token.json'):                                        #Checks for google account sign in for calander events
                creds = Credentials.from_authorized_user_file('token.json', SCOPES)
        
            if not creds or not creds.valid:
                if creds and creds.expired and creds.refresh_token:
                    creds.refresh(Request())
                else:
                    flow = InstalledAppFlow.from_client_secrets_file(
                        'credintials.json', SCOPES)
                    creds = flow.run_local_server(port=0)
                with open('token.json', 'w') as token:
                    token.write(creds.to_json())

            try:
                date_str = data['Start Date'][i].replace('/','-') 
                Edate_str = data['End Date'][i].replace('/','-') 
                date = datetime.strptime(date_str, '%m-%d-%Y')
                Edate = datetime.strptime(Edate_str, '%m-%d-%Y')
# reformat the date string in the year/mm/dd format
                new_date_str = date.strftime('%Y-%m-%d')
                Enew_date_str = Edate.strftime('%Y-%m-%d')
                print(new_date_str)
                service = build('calendar', 'v3', credentials=creds)
                event = {
                    'summary': data.Course[i],                                         #sets up a calander event based on the class name, location and time
                    'location': data.Location[i],
                    'start': {
                        'dateTime': new_date_str+'T'+newdat[1][i]+':00',
                        'timeZone': 'Etc/GMT-3',
                    },
                    'end': {
                        'dateTime': Enew_date_str+'T'+newdat[2][i]+':00',
                        'timeZone': 'Etc/GMT-3',
                    },
                    'recurrence': [
                        'RRULE:FREQ=WEEKLY;COUNT=14'                    #repeats the instance of every class for 14 weeks
                    ],
                    'reminders': {
                        'useDefault': False,
                        'overrides': [
                        {'method': 'popup', 'minutes': 15},
                        ],
                    },
                    }

                event = service.events().insert(calendarId='primary', body=event).execute()              

            except HttpError as error:
                print('An error occurred: %s' % error)

i =0;
while i != size:
    main(i, data, newdat)
    i = i+1;

```
